#### ---> Now we have to merge the two different data in one, for that we can use the Merge function of Pandas.
* In our Two dataset we have common column Instant, so we use Instant to merge the two different dataset
* We can add the dataset to the left or right side of the dataset, here I have use Right to add the dataset2 at the right side of the dataset1
* to_string() of Pandas used to print all the Rows and Columns of dataset 

```python
merged_data = pd.merge(filtered_1, filtered_2, on = "instant", how = "right")
print(merged_data.to_string())
```

* Take the data as Rental bike for bettter understanding


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Group by hour and sum the total counts
hourly_counts = merged_data.groupby('hr')['cnt'].sum().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=hourly_counts, x='hr', y='cnt', palette="Blues_d")

plt.title("Total Bike Rentals by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Rentals (cnt)")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

### The MASTER DATASET
#### ---> The Master dataset is made after filtering the 3 different dataset 
#### ---> Here I have used the Concate function of Pandas Library to concate the Two different dataset (i.g Merged dataset which made from dataset1 & dataset2 and the filtered dataset 3)


```python
master_data = pd.concat([merged_data, filtered_3], ignore_index=True)
master_data
```

#### ---> From the Master data we are going to find the Skewness , Kurtosis and the Outliers of the data
* Using the Numpy and Scipy.stats we can do the different calculation, Matplotlib and Seaborn used for making the plot
* Here the Skewness, Kutrosis and Outliers calculated for all the different columns
* BoxPlot made for the clear visualization of Outliers
* For loop used for iterating each columns of having different data


```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

# Numerical columns to analyze
num_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

# Loop through each numerical column
for col in num_cols:
    data = master_data[col].dropna()
    
    # Skewness and Kurtosis
    sk = skew(data)
    kt = kurtosis(data)
    print(f"\nColumn: {col}")
    print(f"  Skewness: {sk:.4f}")
    print(f"  Kurtosis: {kt:.4f}")
    
    # IQR and Outlier Detection
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    print(f"  IQR: {IQR:.4f}")
    print(f"  Number of outliers: {outliers.count()}")
    
    # Boxplot
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

#### ==> TO know the Correlation of Average Temperature during the Weekday and Hours HeatMap is the best
* Take the Average of Temperature by Hour and Weekday during whole season
* Colorbar beside the Plot useful to know the exect temprature, Blue colur used for low Temperature while Red color used for high Temperature

```python
pivot_temp = master_data.pivot_table(index='weekday', columns='hr', values='temp', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_temp, cmap="coolwarm", annot=False)
plt.title("Average Temperature by Hour and Weekday")
plt.xlabel("Hour")
plt.ylabel("Weekday (Starting from Sunday as 0)")
plt.show()
```
